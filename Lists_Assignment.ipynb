* Create a list named as input_list where the elements are "Abby",1,45,"James",2,"John",1,97.25,"Allan",4,83.50,"Lee",2

```python
input_list = ["Abby",1,45,"James",2,"John",1,97.25,"Allan",4,83.50,"Lee",2]
```

* **Q1:** Add one element 55.10 into the input_list

```python
# Q1
input_list.append(55.10)
print(input_list)
```

* **Q2:** Add element 65.50 at 5th index of the input_list

```python
# Q2
input_list.insert(5,65.50)
print(input_list)
```

* **Q3:** Add elements "Mike",3,75.50 at the end of the input_list

```python
# Q3
input_list.extend(["Mike",3,75.50])
print(input_list)
```

* **Q4:** Show the index position of "Mike" from input_list

```python
# Q4
print(input_list.index("Mike"))
```

* **Q5:** Modify element 45 with 54 of input_list

```python
# Q5
index_45 = input_list.index(45)
input_list[index_45] = 54
print(input_list)
```

* **Q6:** extract all the element from 7th position to before the index position of "Lee" from input_list

```python
# Q6
print(input_list[6:input_list.index("Lee")])
```

* **Q7:** Count how many time 2 Occurred in input_list

```python
# Q7
print(input_list.count(2))
```

* **Q8:** Remove the elements from 10th to 15th position of the input_list

```python
# Q8
del input_list[9:15]
print(input_list)
```

* **Q9:** Add item 7 after 6 in the following Python List<Br>
Given List:<Br>
list_1 = [1, 2, [3, 4, [5, 6], 5], 3, 4]<Br>
Expected output:<Br>
[1, 2, [3, 4, [5, 6, 7], 5], 3, 4]

```python
# Q9
list_1 = [1, 2, [3, 4, [5, 6], 5], 3, 4]
list_1[2][2].append(7)
print(list_1)
```

* **Q10:** Given a nested list extend it by adding the sub list ["k", "l", "m"] in such a way that it will look like the following list<br>
Given List:<Br>
list_1 = ["a", "b", ["c", ["d", "e", ["f", "g"], "h"], "i","j"], "m", ["n",["o","p","q"],"r","s"],"t"]<Br>
Sub List to be added = ["k", "l", "m"]<Br>
Expected output:<Br>
['a', 'b', ['c', ['d', 'e', ['f', 'g'], 'h'], 'i', 'j', 'k', 'l', 'm'], 'm', ['n', ['o', 'p', 'q'], 'r', 's'], 't']

```python
# Q10
list_1 = ['a', 'b', ['c', ['d', 'e', ['f', 'g'], 'h'], 'i', 'j'], 'm', ['n', ['o', 'p', 'q'], ['r', 's'], 't']]

list_1[2].extend(['k', 'l'])
list_1.insert(3, 'm')

print(list_1)
```
